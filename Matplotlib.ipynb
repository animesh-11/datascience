```python
import matplotlib.pyplot as plt
```

```python
plt.plot([2, 4, 6, 4])
plt.ylabel("Numbers")
plt.xlabel("Indices")
plt.title('My Plot')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.ylabel("Squares")
plt.xlabel("Numbers")
plt.title('My Plot')
plt.grid()
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.grid()
plt.show()
```

```python
import numpy as np
t = np.arange(0., 5., 0.2)

#blue dashes, red squares , green triangles
plt.plot(t, t**2, 'b--', label='^2')
plt.plot(t, t**2.2, 'rs', label='^2.2')
plt.plot(t, t**2.5, 'g', label='^2.5')
plt.grid()
plt.legend() # add legend based on line labels
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y, linewidth=10.0)
plt.show()
```

```python
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
x2 = [1, 2, 3, 4]
y2 = [2, 8, 6, 8]
lines = plt.plot(x1, y1, x2, y2)

#use keywor args
plt.setp(lines[0], color='r', linewidth=2.0)

# or MATLAB style string value pairs
plt.setp(lines[1], 'color', 'g', 'linewidth', 2.0)
         
plt.grid()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2* np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
#The subplot() command specifies numrows, numcols
#fignum where fignum from 1 to numrows*numcols.
plt.subplot(211) # 2 rows , 1st figure, 1st subplot
plt.grid()
plt.plot(t1, f(t1), 'b-')

plt.subplot(212) # 2 rows , 1st figure, 2nd subplot
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
plt.figure(1)  # the first figure
plt.subplot(211) #the first subplot in the first figure
plt.plot([1, 2, 3]) 
plt.subplot(212) #the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2) # the second figure
plt.plot([4, 5, 6]) #creates a subplot(111) by default

plt.figure(1)  # figure 1 current subplot(211)
plt.subplot(211) #make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') #subplot 211 title
plt.show()
```
